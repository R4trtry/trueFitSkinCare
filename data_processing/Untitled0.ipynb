```python
import numpy
import pandas as pd
n = random.randint(1177)
sim_rating = np.tril(np.loadtxt("sim_product2.csv",delimiter=',')).flatten()
sim_review = np.tril(np.loadtxt("sim_product_review2.csv",delimiter=',')).flatten()
sim_quick = np.tril(np.loadtxt("sim_product_quicktake2.csv",delimiter=',')).flatten()

import MySQLdb
def sql2df(sql):
    con = MySQLdb.connect('localhost', 'root', '', 'insight')
    with con:
        cur = con.cursor(MySQLdb.cursors.DictCursor)
        df = pd.read_sql_query(sql,con)
    return df
product_list = sql2df('select product_id, category, brand, price, product_name from Product order by category, brand;')
```

```python
price = np.array(product_list['price'].tolist())
category = product_list['category'].tolist()
brand = product_list['brand'].tolist()
brand_mat = []
price_mat = []
cat_mat = []
for i in range(1177):
    price_mat.append(price-price[i])
    cat_mat.append([1 if category[i]==c else 0 for c in category])
    brand_mat.append([1 if brand[i]==c else 0 for c in brand])
    
price_diff = np.tril(price_mat).flatten()
color = np.tril(brand_mat).flatten()
color = ['b' if c else 'g' for c in color]
```

```python
d = {'sim_rating':sim_rating, 'sim_review':sim_review, 'sim_quick':sim_quick, 'price_diff':price_diff, 'color':color}
df = pd.DataFrame(d)
df = df[df[['sim_rating','sim_review','sim_quick']].sum(1)>0.01]
df = df.sort(['sim_rating'], ascending=True)
```
